```python
import os
import re
from khmernormalizer import khmer_normalize
from khmernltk import word_tokenize

# Folders containing your input files
input_folders = [""
"data-mptc", 
"text-mptc"
""]  # replace with your folders

# Output folder
output_folder = "datasets/"
os.makedirs(output_folder, exist_ok=True)

# Output combined file path
combined_file = os.path.join(output_folder, "all_cleaned_words.txt")

all_words = []

# Process each folder
for folder_path in input_folders:
    files = [f for f in os.listdir(folder_path) if f.endswith(".txt")]
    
    for input_file in files:
        input_path = os.path.join(folder_path, input_file)
        
        with open(input_path, "r", encoding="utf-8") as f:
            for line in f:
                text = line.strip()
                
                # Remove emojis & unwanted symbols (keep Khmer unicode + space)
                text = re.sub(r"[^\u1780-\u17FF\u200B\u0020]+", "", text)
                
                # Normalize Khmer text
                text = khmer_normalize(text)
                
                # Tokenize into words
                tokens = word_tokenize(text)
                
                # Add each non-empty word
                all_words.extend([word for word in tokens if word.strip()])
        
        print(f"Processed {folder_path}/{input_file}")

# Save all words into one combined file (one word per line)
with open(combined_file, "w", encoding="utf-8") as f:
    for word in all_words:
        f.write(word + "\n")

print(f"All files cleaned, tokenized, and saved in '{combined_file}'")
```
